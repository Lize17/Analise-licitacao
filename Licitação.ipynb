# Importando as Bobliotecas 

```python
import pandas as pd
import os
```

```python
arquivos = os.listdir()

licitacao = pd.read_excel('Empresa A.xlsx')

licitacao = licitacao.rename(columns={'Total' : 'Empresa A'})

licitacao = licitacao.drop(['Valor Unitário'], axis=1)
#display(licitacao)

empresas = arquivos[18:25]
#print(empresas)
for empresa in empresas:
 proposta = pd.read_excel(empresa)
 licitacao[empresa[:-5]] = proposta['Total'].copy()
display(licitacao)
```

# Verificando Valores Totais

```python
for coluna in licitacao.columns[3:]:
   total = licitacao[coluna].sum()
   print(f'{coluna} : R$ {total:,.2f}')
```

# Análise inicial e possíveis ações:

2. **Empresa B** possui um total significativo de R$ 267,971.00, que é consideravelmente menor que o da Empresa F, mas ainda substancial.

Empresas que recebem mais recursos devem ser monitoradas de perto para garantir que estão cumprindo com os requisitos contratuais e oferecendo um bom retorno sobre o investimento.

- **Negociação de Contratos:** Considerar a possibilidade de renegociar contratos com as empresas que receberam valores elevados para buscar melhores condições financeiras ou serviços adicionais que possam agregar mais valor à licitação.

- **Diversificação de Fornecedores:** Avaliar a possibilidade de diversificar os fornecedores para reduzir a dependência de uma única empresa, especialmente aquelas que recebem os maiores montantes.

- **Revisão de Critérios de Seleção:** Rever os critérios de seleção utilizados na licitação para garantir que estão promovendo uma competição justa e selecionando os fornecedores mais adequados com base em critérios objetivos.

- **Transparência e Prestação de Contas:** Manter um alto padrão de transparência na gestão dos recursos públicos, garantindo que todas as decisões e alocações sejam bem documentadas e justificadas.


# Resultado "ótimo": 

```python
licitacao['Mínimo'] = licitacao.drop(['Item','Descrição do Item','Quantidade'],axis=1).min(axis=1)
display(licitacao)

minimo = licitacao['Mínimo'].sum()
print(minimo)
```

# Criando uma planilha de desvios por fornecedor:

```python
for empresa in licitacao.columns[3:-1]:
    desvio = licitacao [['Item','Descrição do Item','Quantidade', 'Empresa A', 'Mínimo']]
    desvio['Delta'] = desvio['Empresa A'] - desvio['Mínimo']
    desvio.to_excel('Desvio - ' + empresa + '.xlsx', index = False)
```
